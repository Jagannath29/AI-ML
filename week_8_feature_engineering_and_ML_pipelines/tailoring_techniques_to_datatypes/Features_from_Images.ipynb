__Introduction__

In-text analysis, a word is easily extractable. A word is a basic semantic unit in a language. Images and audio, on the other hand, are recorded as digital
pixels or waveforms. These contain much less semantic information than an atom —a word—of text data. Therefore, feature extraction and engineering are much more challenging on image and audio than on text.

We need to suitably encode images for machine learning.The simplest approach is to use images pixel values.



**Pixel**

The screen is represented by a 2D array of locations called pixels. The pixel values, denote the intensity or brightness of the pixel. Smaller numbers (closer to zero) represent black, and larger numbers (closer to 255) denote white.


A digital image is made up of rows and columns of pixels.  A pixel
can be identified by  coordinates giving column number and the row number.

 _For
example, the pixel (3,5) would lie in column  3 and row 5 in a coordinate_




Column  and Row numbers identify a pixel. They don't identify a point. A pixel contains many points;
mathematically, it contains an infinite number of points.

**Is Pixel always suitable?**

Let’s say our task is
image retrieval: we are given a picture and asked to find similar photos.
Now we need to decide how to represent each image, and how to
measure the differences between them. Can we look at the percentage of
different colors in an image?

> The Figure below shows two pictures having the same shape, and
same color but very different meanings. Therefore color information is probably
not enough to characterize an image.

> Another simple idea is to measure the pixel value differences between images.

1. Resize the images to have the same width and height. A matrix of pixel values represents each image. The color of each pixel (i.e., the RGB
encoding) is now a feature of the image.

2. Measure the
Euclidean distance between the long pixel vectors.
In the real world, individual pixels do not carry enough semantic information, and they are rarely uniform. Hence simple pixels will not be suitable for analysis.




**Note ...**


Previously, Image feature extractors such as   Scale-Invariant Feature Transform [SIFT](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_sift_intro/py_sift_intro.html), and Histograms of Oriented Gradients [HOG](https://lear.inrialpes.fr/people/triggs/pubs/Dalal-cvpr05.pdf) were the

reach of traditional machine learning models by incorporating automatic feature
extraction.

Deep Learning models replace manually defined feature
image extractors with  defined models that automatically learn and
extract features. The manual work is still there,but just abstracted.

# **Scikit-image**

scikit-image (a.k.a. skimage) is a collection of algorithms for image processing and computer vision. The basic image is just a standard NumPy array, which exposes pixel data directly to the user. Hence, many common operations can be achieved using standard NumPy methods for manipulating arrays. A new user can simply the load an image from disk, process that image with image filters, and quickly display the results.

This Section describes how to use scikit-image on various image processing tasks:

**scikit-image and the SciPy ecosystem**

```python
import skimage
from skimage import data  # most functions are in subpackages
```

*Most scikit-image functions take NumPy ndarrays as arguments.*

**Input/output, data types and colorspaces**

```python
from skimage import io #
# Reading from files: skimage.io.imread()
import os
filename = os.path.join(skimage.data_dir, 'camera.png')
camera = io.imread(filename)
```

```python
io.imshow(camera)
```

_Alternatively you can use matplotlib to display image._

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(camera)
plt.show()  # display it
```
