Hyperparameter tuning refers to choosing the optimization settings (hyperparameters) that are selected manually or programmatically to make the machine learning model perform its best, i.e. find the model hyperparameters that yield the best score on the validation metrics. Hyperparameters define the model behaviour and parameters that can be learned. Now, we are going to discuss on problems we encounter during executing hyperparameter tuning.


## **Trusting the defaults**

 Trusting the defaults is like performing no hyperparameter tunings at all. When we perform machine learning operations using libraries, there are already various defaults initialised on the libraries. These defaults can provide a good result in some cases but in most cases they are proved to provide bad results.Since the objective, data types, data volumes and goals changes in every machine learning projects. The defaults hyperparameter values cannot fits all models and all problems and common problem nowadays is not changing the default hyperparameter of the model which isn't always right for your problem. For example, consider that we are traininig a neural network model and our training size is about 2 Millions. Here, in this case if we use the default batch size of 32 then the model will probably take a very large time to converge and may also not provide a good results.  This is one of the problem of hyperparameter tuning and optimization.

In conclusion, we should not use default set of hyperparameters, we should set the hyperparameters on our own. And for this we must have enough knowledge of data we are training with and the model architecture.




## **Using Wrong Metrics**
After completing data preprocessing, training process and  prediction, the next step is to evaluate the performance of the model based on some metric using test datasets. The metrics measure the performance of the model. Suppose, a model provide desirable outcome when evaluating with accuracy_score but can give poor performance if we use different metrics.or suppose, consider a dataset for our machine learning model that consists of a lot of outliers. Using the error functions like: Mean Squarred Errors without proper analysis will give out worst performance results. Moreover, thinking ReLU gives the best performance results compared to the other activation functions and using ReLU on all the neural network layers can also give rise to exploding gradient problems. So, it is necessary to use the right metrics at the right place. Hence, it is necessary to select right metrics to measure the performance of machine learning models.

What if we chose wrong metrics?
Suppose, in Twitter an algrorithm prioritizes a tweet based on how many retweets it receives. Harrassing tweets may receive more retweets due to denouncing the statement. But tweet responses from the victim to the harasser tweets may get likes but less retweets. That's how twitter algrorithms prioritize the harasser tweets over the victim ones and shows how wrong metrics leads to undesirable outcomes. So, hyperparameter optimization might help to choose right metrics depending on your problem.


## **Overfitting**
A good machine learning model can generize any new dataset from problem domain in a proper way. It will able to make prediction on future or unseen dataset. While building machine learning model, we face overfitting which can be responsible for the bad performance. The model is overfitted when we train our model on huge dataset, it started to get trained on outliers and inaccurate entries. and the model performs well on training data and get poor performance on unseen testing data.

Tuning only a single or a couple of hyperparameters may give out a bad results as well. For example while  hyperparameter tuning, tuning only the activation function or the no. of neurons won't laid out a good performance result. We need to tune no. of neurons, layers, activation functions, no. of epochs and learning rate. With this we can obtained a good results.

## **Hand Tuning**
Optimization method is the method in which the next set of hyperparameter are provided  during hyperparameter optimization. The main goal of optimization methods is to find the optimal solution with low computational efforts.

When we use hand tuning method, we change a hyperparameter manually based on current choice of hyperparameters and their performance and train the model again and check the difference in the performance. In this case, we are the optimization method. It is an inefficient way to optimize the model. The main drawbacks are that humans are poor at visualizing high dimensional, and it contain lots of manual work, time required and also inefficient for large models.

## **Grid Search**
In Grid search, we search parameters in a grid like structure specified by combination of possible values of parameters.The main drawbacks of grid search is that, it suffers from crush of dimensionality and it takes a lot of time to try all the combinations. The larger the number of hyperparameters, the larger the time taken. The number of combinations increases drastically when one parameter is increased. Trying all the combinations then becomes infeasible for a large number of hyperparameters. Hence, this method is not generally used for hyperparameter tuning in deep neural networks with lots of hyperparameters.

## **Random Search**
In Random Search, we randomly sample parameters from their possible distribution and use all randomly sampled parameters. We use random search to randomly sample hyperparameters in Deep Learning models.The main problem of random search is unnecessary variance and it randomly sample hyperparameters and uses no intelligence in selecting which combination to try.Although this method is beneficial to find best among the various randomly sampled hyperparameters, we cannot find the best possible hyperparameter.

## Bayesian Optimization
Grid search and Random search for the optimization of the model. Although these methods are slightly better than manual tuning, they are not the best one. One of the major drawbacks of these methods is they don't evaluate results of previous evaluations. As a result, a massive amount of time is spent on evaluating 'bad' hyperparameters. This problem is solved by Bayesian Optimization.Bayesian Optimization is an optimization strategy for optimizing black-box functions. Bayesian Optimization decreases the minimum number of trials required to optimize the function.It reasons about the hyperparameters to find the best ones to evaluate. This significantly decreases the number of searches required. It is flexible and can be used to search for any set of hyperparameters.
